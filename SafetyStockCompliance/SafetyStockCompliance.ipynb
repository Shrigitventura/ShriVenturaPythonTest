```python
##Exception Report Processing from JDE

import pandas as pd

# Define file path
file_path = "C:/Users/SPoudel/Ventura Foods/SC Analytics Team - General/Stan Report Files/File Repository/Safety Stock Compliance/Weekly Run Files/2025/01.28.2025/exception report.xlsx"

# Read Excel file, skipping the first 3 rows (so row index 3 becomes the header)
exception_report = pd.read_excel(file_path, sheet_name="Sheet1", skiprows=3, engine="openpyxl")

# Reset index (optional, but ensures clean row numbers)
exception_report = exception_report.reset_index(drop=True)

# Replace spaces with underscores in column names
exception_report.columns = [
    "B_P", "ItemNo", "Buyer", "Planner", "Supplier_No", "Payee_Number", "MPF_or_Line", "Order_Policy_Code", "Order_Policy_Value", "Plan_Code", "Fence_Rule", "Plan_Fence_Days",
    "Msg_Display_Fence", "Freeze_Fence", "Leadtime_Days", "Reorder_MIN", "Reorder_MAX", "Reorder_Multiple", "Safety_Stock", "Reorder_Point", "Reorder_Qty", "Avg_Demand_Weeks",
    "Formula_Type", "Sort_Code", "Schedule_Group", "Model", "Description", "UOM", "PL_QTY", "Planning_Formula", "Costing_Formula"
]

# Create a new 'ref' column by concatenating 'B_P' and 'ItemNo' with an underscore
exception_report["ref"] = exception_report["B_P"].astype(str) + "_" + exception_report["ItemNo"].astype(str)

# Remove duplicate rows based on the 'ref' column
exception_report = exception_report.drop_duplicates(subset=["ref"])

# Replace NA values in the 'Planner' column with 0 then convert to float to accept 0's and then to int to conver to whole number and then str to convert numberts without decimal to string
exception_report["Planner"] = exception_report["Planner"].fillna(0).astype(float).astype(int).astype(str)


# Display the cleaned DataFrame
print(exception_report.head())
```
